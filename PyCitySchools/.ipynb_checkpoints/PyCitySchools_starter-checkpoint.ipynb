```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data_complete["school_name"].nunique()
total_schools
```

```python
#Calculate the total number of students
total_students = school_data_complete["Student ID"].nunique()
total_students
```

```python
#Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average maths score
average_maths_score = school_data_complete["maths_score"].mean()
average_maths_score
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#Calculate the percentage of students with a passing maths score (50 or greater)
passing_maths = school_data_complete[school_data_complete["maths_score"] >= 50]
over_50_maths = passing_maths["maths_score"].count() / total_students * 100
over_50_maths
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater)
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 50]
over_50_reading = passing_reading["reading_score"].count() / total_students * 100
over_50_reading
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
passing_both = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
overall_passing = len(passing_both) / len(school_data_complete) * 100
overall_passing
```

```python
#Create a dataframe to hold the above results
government_area_df = pd.DataFrame([{"Total Schools": total_schools, "Total Students": total_students, "Total Budget": total_budget, 
                    "Average Maths Score": average_maths_score, "Average Reading Score": average_reading_score,
                   "% Passing Maths": over_50_maths, "% Passing Reading": over_50_reading, "% Overall Passing": overall_passing}])
government_area_df.head()
```

```python
#Optional: give the displayed data cleaner formatting
government_area_df["Total Budget"] = government_area_df["Total Budget"].map("${:,.2f}".format)
government_area_df["% Passing Maths"] = government_area_df["% Passing Maths"].map("{:.6f}%".format)
government_area_df["% Passing Reading"] = government_area_df["% Passing Reading"].map("{:.6f}%".format)
government_area_df["% Overall Passing"] = government_area_df["% Overall Passing"].map("{:.6f}%".format)
government_area_df.head()
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Creating grouped data series for average maths score by year
year_9_maths = student_data.loc[student_data["year"] == 9, :].groupby("school_name")["maths_score"].mean()
year_10_maths = student_data.loc[student_data["year"] == 10, :].groupby("school_name")["maths_score"].mean()
year_11_maths = student_data.loc[student_data["year"] == 11, :].groupby("school_name")["maths_score"].mean()
year_12_maths = student_data.loc[student_data["year"] == 12, :].groupby("school_name")["maths_score"].mean()
```

```python
average_maths_scores_by_year = pd.DataFrame({"Year 9": year_9_maths, "Year 10": year_10_maths, "Year 11": year_11_maths, "Year 12": year_12_maths})
average_maths_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
schools_grouped_df.dtypes
```

```python
bins = [0, 585, 630, 645, 680]
group_labels = ["<$585", "$585-630", "$630-645", "$645-680" ]
schools_grouped_df["Spending Ranges (Per Student)"] = pd.cut(schools_grouped_df["Per Student Budget"], bins, labels = group_labels)
schools_grouped_df
```

```python
spending_maths_scores = schools_grouped_df.groupby("Spending Ranges (Per Student)")["Average Maths Score"].mean()
spending_reading_scores = schools_grouped_df.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()
spending_passing_maths = schools_grouped_df.groupby("Spending Ranges (Per Student)")["% Passing Maths"].mean()
spending_passing_reading = schools_grouped_df.groupby("Spending Ranges (Per Student)")["% Passing Reading"].mean()
spending_passing_both = schools_grouped_df.groupby("Spending Ranges (Per Student)")["% Overall Passing"].mean()
```

```python
scores_by_school_spending = pd.DataFrame({"Average Maths Score": spending_maths_scores, "Average Reading Score": spending_reading_scores, "% Passing Maths": spending_passing_maths, "% Passing Reading": spending_passing_reading, "% Overall Passing": spending_passing_both})
scores_by_school_spending
```

```python
scores_by_school_spending["Average Maths Score"] = scores_by_school_spending["Average Maths Score"].map("{:.2f}".format)
scores_by_school_spending["Average Reading Score"] = scores_by_school_spending["Average Reading Score"].map("{:.2f}".format)
scores_by_school_spending["% Passing Maths"] = scores_by_school_spending["% Passing Maths"].map("{:.2f}".format)
scores_by_school_spending["% Passing Reading"] = scores_by_school_spending["% Passing Reading"].map("{:.2f}".format)
scores_by_school_spending["% Overall Passing"] = scores_by_school_spending["% Overall Passing"].map("{:.2f}".format)
scores_by_school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 1000, 2000, 5000]
group_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
schools_grouped_df["School Size"] = pd.cut(schools_grouped_df["Total Students"], bins, labels = group_labels)
schools_grouped_df
```

```python
size_maths_scores = schools_grouped_df.groupby("School Size")["Average Maths Score"].mean()
size_reading_scores = schools_grouped_df.groupby("School Size")["Average Reading Score"].mean()
size_passing_maths = schools_grouped_df.groupby("School Size")["% Passing Maths"].mean()
size_passing_reading = schools_grouped_df.groupby("School Size")["% Passing Reading"].mean()
size_passing_overall = schools_grouped_df.groupby("School Size")["% Overall Passing"].mean()
```

```python
scores_by_school_size = pd.DataFrame({"Average Maths Score": size_maths_scores, "Average Reading Score": size_reading_scores, "% Passing Maths": size_passing_maths, "% Passing Reading": size_passing_reading, "% Overall Passing": size_passing_overall})
scores_by_school_size
```
